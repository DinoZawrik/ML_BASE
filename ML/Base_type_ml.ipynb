## **Основные типы машинного обучения**

* **Три основных типа машинного обучения:**

1.  **Обучение с учителем (Supervised Learning):** 👨‍🏫📚
    *   **Суть:**  Модель обучается на размеченных данных, то есть на данных, где для каждого примера есть известный правильный ответ (целевая переменная, метка).  Цель – научиться предсказывать этот правильный ответ для новых, невиданных ранее данных.
    *   **Типичные задачи:**
        *   **Классификация:**  Предсказание *категории* объекта (например, определить, спам ли письмо, классифицировать изображение как кошку или собаку).
        *   **Регрессия:**  Предсказание *числового значения* (например, предсказать цену дома, спрогнозировать температуру воздуха).

2.  **Обучение без учителя (Unsupervised Learning):** 🕵️‍♂️🌍
    *   **Суть:**  Модель обучается на неразмеченных данных, то есть на данных без известных правильных ответов.  Цель – найти структуру в данных, выделить скрытые закономерности, группировать похожие объекты вместе, уменьшить размерность данных и т.д.
    *   **Типичные задачи:**
        *   **Кластеризация:**  **Разбиение данных на группы (кластеры) похожих объектов (например, сегментация клиентов на группы со схожими характеристиками).
        *   **Понижение размерности:**  Уменьшение количества признаков в данных, сохраняя при этом важную информацию (например, сжатие изображений, визуализация многомерных данных на плоскости).
        *   **Поиск аномалий:  **Выявление необычных или аномальных объектов, которые отличаются от основной массы данных (например, обнаружение мошеннических транзакций).

3.  **Обучение с подкреплением (Reinforcement Learning):** 🎮🤖
    *   **Суть:**  Модель обучается принимать последовательность решений в динамической среде, чтобы достичь определенной цели.  Обучение происходит за счет получения обратной связи в виде наград или штрафов за каждое действие.  Нет "правильных ответов" в явном виде, есть только цель – максимизировать награду.
    *   **Типичные задачи:**
        *   **Игры:**  Обучение компьютерных игр (например, AlphaGo, шахматы, Atari).
        *   **Робототехника:**  Управление роботами, обучение их навигации, манипуляция объектами.
        *   **Рекомендательные системы:**  Персонализация рекомендаций для пользователей на основе их взаимодействий и обратной связи.
        *   **Оптимизация:**  Оптимизация бизнес-процессов, управление ресурсами, торговля на бирже.

* **Основные этапы разработки ML-модели:**

1.  **Сбор данных (Data Collection):** 📊📥 На этом этапе мы собираем данные, которые будут использоваться для обучения нашей модели.

2.  **Предобработка данных (Data Preprocessing):** 🧹✨ Собранные данные редко бывают идеальными –  в них могут быть ошибки, пропуски, неверный формат, выбросы, шум.  Предобработка – это очистка, преобразование и подготовка данных к обучению модели.  Сюда входит обработка пропусков, удаление дубликатов, преобразование типов данных, масштабирование признаков и многое другое.

3.  **Инженерия признаков (Feature Engineering):** 🛠️⚙️ Признаки (features) – это входные переменные, которые модель использует для обучения и предсказаний.  Инженерия признаков – это процесс отбора, создания и преобразования признаков, чтобы улучшить качество и производительность модели.  Сюда входит выбор наиболее важных признаков, создание новых признаков на основе существующих, преобразование признаков (например, кодирование категориальных признаков, создание полиномиальных признаков).

4.  **Выбор модели (Model Selection):** 🧠🧰 Существует множество различных алгоритмов машинного обучения (линейная регрессия, логистическая регрессия, деревья решений, нейронные сети, SVM и т.д.).  Выбор подходящей модели зависит от типа задачи (классификация, регрессия, кластеризация и т.д.) и характеристик данных.  На этом этапе мы выбираем наиболее перспективные модели для нашей задачи.

5.  **Обучение модели (Model Training):** 🏋️‍♀️💪 Обучение модели – это процесс настройки параметров выбранной модели на основе обучающих данных.**  Модель "учится" на данных, находя закономерности и зависимости, чтобы потом делать верные предсказания на новых данных.  Для обучения используются алгоритмы оптимизации, например, градиентный спуск, который мы будем изучать позже.

6.  **Оценка модели (Model Evaluation):** 📏🧐 После обучения модель нужно оценить, насколько хорошо она работает.  Для этого используются метрики оценки качества (accuracy, precision, recall, F1-score, MSE, MAE, и т.д.).  Оценка обычно проводится на отложенной тестовой выборке, которую модель не видела во время обучения, чтобы получить объективную оценку ее способности к обобщению на новые данные.

7.  **Настройка гиперпараметров (Hyperparameter Tuning/Optimization)**: ⚙️🔧 У большинства ML-моделей есть гиперпараметры – параметры, которые не настраиваются в процессе обучения, а задаются заранее (например, глубина дерева решений, коэффициент регуляризации в линейной регрессии).  Настройка гиперпараметров – это процесс подбора оптимальных значений гиперпараметров, чтобы максимизировать качество модели.  Для этого используются разные методы, например, перебор по сетке (Grid Search), случайный поиск (Random Search), байесовская оптимизация.

8.  **Развертывание модели (Model Deployment):** 🚀🛰️ Развертывание модели – это процесс внедрения обученной модели в реальную среду, чтобы ее можно было использовать для решения практических задач.  Это может быть веб-приложение, мобильное приложение, API, встраивание в другое программное обеспечение и т.д.

9.  **Мониторинг и поддержка модели (Model Monitoring and Maintenance):** 📊🛠️ После развертывания модель нужно постоянно мониторить и поддерживать.  Со временем качество модели может ухудшаться из-за изменения данных, "дрейфа данных" (data drift), появления новых данных, изменения бизнес-требований.  Мониторинг позволяет отслеживать качество модели в продакшене, выявлять проблемы и принимать меры по ее обновлению, переобучению или замене.

#### **Метрики**    
Для оценки моделей используются **метрики оценки качества**. Метрика - это числовая величина, которая количественно измеряет качество работы модели.

##### **Метрики для задач РЕГРЕССИИ.**

**Метрики регрессии** используются для оценки моделей, которые **предсказывают *числовые* значения**.  Например, предсказание цены дома, температуры, спроса на товар.

1.  **Mean Squared Error (MSE) – Среднеквадратичная ошибка:** 📏
    *   **Интуитивно:**  Среднее значение *квадратов* ошибок (разностей) между *предсказанными* и *реальными* значениями.
    *   **Как считается:**  Для каждой точки данных вычисляется разность между предсказанным значением (модель предсказала) и реальным значением (на самом деле было).  Эта разность возводится **в квадрат** (чтобы избавиться от знака и усилить большие ошибки).  Затем все эти **квадраты ошибок усредняются**.
    *   **Что показывает:**  MSE показывает средний размер "квадратичной" ошибки модели.  Чем меньше MSE, тем лучше модель.  MSE очень чувствительна к *большим ошибкам*, так как они возводятся в квадрат.  Поэтому MSE хорошо подходит, когда важно избегать *больших* ошибок.
    *   **Единицы измерения:**  Единицы измерения MSE – это квадрат единиц измерения целевой переменной.  Например, если целевая переменная – цена в долларах, то MSE будет в "долларах в квадрате", что не всегда интуитивно понятно.
    *   **Формула MSE:**    
    $$\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$  
    * Где:
        * n - количество наблюдений,
        * $y_i$ - истинное значение,
        * $\hat{y_i}$ - предсказанное значение.

2.  **Mean Absolute Error (MAE) – Средняя абсолютная ошибка:** 🔢
    *   **Интуитивно:**  Среднее значение *абсолютных значений* ошибок (разностей) между *предсказанными* и *реальными* значениями.
    *   **Как считается:**  Похоже на MSE, но разность между предсказанным и реальным значением берется по *абсолютному значению* (модулю), а не возводится в квадрат.  Затем эти абсолютные ошибки усредняются.
    *   **Что показывает:**  MAE показывает средний размер "абсолютной" ошибки модели.  Чем меньше MAE, тем лучше модель.  MAE менее чувствительна к *выбросам* и *большим ошибкам*, чем MSE, так как абсолютное значение "сглаживает" большие ошибки, не так сильно их усиливая, как квадрат.  MAE хорошо подходит, когда важно получить *среднюю* ошибку в *исходных единицах измерения*.
    *   **Единицы измерения:**  Единицы измерения MAE – совпадают с единицами измерения целевой переменной.  Например, если целевая переменная – цена в долларах, то MAE будет в "долларах", что более интуитивно понятно, чем "доллары в квадрате" у MSE.
    *   **Формула MAE:**
    $$\frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$$

    * Где:
        - n — количество наблюдений,
        - $y_i$ — истинное значение,
        - $\hat{y}_i$ — предсказанное значение.


3.  **Mean Absolute Percentage Error (MAPE) – Средняя абсолютная процентная ошибка:** ％
    *   **Интуитивно:**  Средний процент ошибки модели относительно *реальных* значений.
    *   **Как считается:**  Для каждой точки данных вычисляется относительная ошибка –  абсолютная разность между предсказанным и реальным значением, деленная на *реальное* значение.  Затем эта относительная ошибка выражается в процентах (умножается на 100%).  И эти процентные ошибки усредняются.
    *   **Что показывает:**  MAPE показывает средний процент ошибки модели.  Чем меньше MAPE, тем лучше модель.  MAPE легко интерпретировать –  например, MAPE = 5% означает, что в среднем модель ошибается на 5% от реального значения.  MAPE удобна для сравнения моделей на разных шкалах измерений, так как это безразмерная величина (процент).  Важно:  MAPE нельзя использовать, если в данных есть *нулевые* или *очень близкие к нулю* реальные значения, так как деление на ноль или очень маленькое число приведет к бесконечным или очень большим значениям MAPE.
    *   **Единицы измерения:**  Проценты (%) – безразмерная величина.

**Интуитивное понимание "отклонения" и "ошибки":**

Все эти метрики (MSE, MAE, MAPE) по сути измеряют **"отклонение" или "ошибку"** предсказаний модели от реальных значений.  Чем **меньше** это отклонение, тем **лучше** модель.  Разница – в том, **как именно это "отклонение" измеряется и усредняется**:

*   **MSE – "квадратичное отклонение" (сильно штрафует за большие ошибки)**
*   **MAE – "абсолютное отклонение" (более устойчива к выбросам)**
*   **MAPE – "процентное отклонение" (легко интерпретировать, но осторожно с нулями)**

##### **Метрики для задач КЛАССИФИКАЦИИ.**

**Основные метрики классификации:**

1.  **Accuracy (Точность)**
2.  **Confusion Matrix (Матрица ошибок)** - это основа для многих других метрик
3.  **Precision (Точность)**
4.  **Recall (Полнота)**
5.  **F1-score (F1-мера)**
6.  **ROC AUC (Площадь под ROC-кривой)**
7.  **PR AUC (Площадь под PR-кривой)**

---

**1. Accuracy (Точность)**

**Accuracy** –  самая простая и, возможно, самая интуитивно понятная метрика классификации.  Она показывает **долю *верно классифицированных* объектов от *общего количества* объектов.**

**Формула Accuracy:**

$ {Accuracy} = \frac{\text{Количество верных}}{\text{Общее количество}}$ 

Или, в терминах **матрицы ошибок (Confusion Matrix):**


$ {Accuracy} = \frac{\text{TP + TN}}{\text{TP + TN + FP + FN}}$ 


Где:

*   **TP (True Positives) – Истинно-положительные срабатывания:**  Количество объектов, которые модель **верно классифицировала как *положительный класс*** (например, верно определила, что письмо – спам).
*   **TN (True Negatives) – Истинно-отрицательные срабатывания:**  Количество объектов, которые модель **верно классифицировала как *отрицательный класс*** (например, верно определила, что письмо – не спам).
*   **FP (False Positives) – Ложно-положительные срабатывания (ошибки I рода):**  Количество объектов, которые модель **ошибочно классифицировала как *положительный класс***, хотя на самом деле они относятся к *отрицательному классу* (например, ошибочно определила, что письмо – спам, хотя оно не спам – "ложная тревога").
*   **FN (False Negatives) – Ложно-отрицательные срабатывания (ошибки II рода):**  Количество объектов, которые модель **ошибочно классифицировала как *отрицательный класс***, хотя на самом деле они относятся к *положительному классу* (например, ошибочно определила, что письмо – не спам, хотя оно спам – "пропуск цели").

Accuracy показывает, насколько часто модель *в целом* правильно угадывает классы объектов.  Чем выше Accuracy, тем лучше модель.  **Идеальное значение Accuracy – 1.0 (или 100%)**, что означает, что модель идеально классифицирует все объекты (что на практике почти никогда не достигается).

Несмотря на свою простоту и интуитивность, Accuracy имеет *серьезные ограничения*, особенно в задачах с *несбалансированными классами (imbalanced datasets)*. Несбалансированные классы – это когда количество объектов разных классов *сильно различается*.  Например, в задаче обнаружения редких заболеваний, больных людей может быть *очень мало* (например, 1%), а здоровых – *очень много* (99%).  Или в задаче обнаружения мошеннических транзакций, мошеннических транзакций тоже может быть *гораздо меньше*, чем обычных. В случае несбалансированных классов, модель может достигать *высокой Accuracy*, просто *всегда предсказывая *мажоритарный класс* (тот, которого большинство), и при этом быть *абсолютно бесполезной* для *миноритарного класса*** (того, который нас как раз интересует!).

**2. Confusion Matrix (Матрица ошибок)** –  это не метрика, а скорее *таблица*, которая показывает *все возможные исходы классификации*, сравнивая *истинные* и *предсказанные* классы для набора данных.  Она помогает детально проанализировать ошибки классификации и понять, в каких классах модель ошибается чаще, а в каких – реже, и какие *типы* ошибок преобладают.

**Структура Confusion Matrix для бинарной классификации:**

Для бинарной классификации Confusion Matrix – это **квадратная матрица 2x2**:

|                | **Предсказанный класс: Положительный** | **Предсказанный класс: Отрицательный** |
| :------------- | :------------------------------------: | :-------------------------------------: |
| **Истинный класс: Положительный** | **TP (True Positives)**             | **FN (False Negatives)**             |
| **Истинный класс: Отрицательный** | **FP (False Positives)**             | **TN (True Negatives)**             |

**Разберем ячейки матрицы:**

*   **TP (True Positives) – Истинно-положительные срабатывания:**  Объекты, которые на самом деле относятся к *положительному классу*, и модель верно предсказала, что они относятся к *положительному классу*.  **Верные срабатывания для положительного класса.**
*   **TN (True Negatives) – Истинно-отрицательные срабатывания:**  Объекты, которые на самом деле относятся к *отрицательному классу*, и модель верно предсказала, что они относятся к *отрицательному классу*.  **Верные срабатывания для отрицательного класса.**
*   **FP (False Positives) – Ложно-положительные срабатывания (Ошибки I рода):**  Объекты, которые на самом деле относятся к *отрицательному классу*, но модель ошибочно предсказала, что они относятся к *положительному классу*.  **Ложные тревоги, ошибки "пере-бдительности".**
*   **FN (False Negatives) – Ложно-отрицательные срабатывания (Ошибки II рода):**  Объекты, которые **на самом деле относятся к *положительному классу***, но модель **ошибочно предсказала, что они относятся к *отрицательному классу***.  **Пропуски цели, ошибки "недо-бдительности".**

**Как "читать" Confusion Matrix?**

Confusion Matrix показывает **распределение предсказаний модели по отношению к истинным классам**.  Идеальная модель будет иметь **большие значения на *главной диагонали* (TP и TN)** и **маленькие значения *вне диагонали* (FP и FN)**.  **Чем больше значения TP и TN и чем меньше FP и FN – тем лучше модель.**

**Пример Confusion Matrix:**

Предположим, для той же задачи классификации спама (100 писем), Confusion Matrix выглядит так:

|                | Предсказанный: Спам | Предсказанный: Не-спам |
| :------------- | :-----------------: | :--------------------: |
| Истинный: Спам |        **TP = 20**       |        **FN = 5**        |
| Истинный: Не-спам |       **FP = 5**        |       **TN = 70**       |

**Интерпретация:**

*   **TP = 20:**  Модель верно определила 20 писем как спам (из общего числа спам-писем).
*   **FN = 5:**  Модель пропустила 5 спам-писем, ошибочно классифицировав их как не-спам.
*   **FP = 5:**  Модель ошибочно пометила 5 не-спам писем как спам ("ложные срабатывания").
*   **TN = 70:**  Модель верно определила 70 писем как не-спам (из общего числа не-спам писем).

**Глядя на Confusion Matrix, мы можем увидеть:**

*   **Общее количество верных предсказаний (Accuracy):**  TP + TN = 20 + 70 = 90.  Accuracy = 90/100 = 90%.
*   **Какие типы ошибок преобладают:**  В данном примере FP = 5 и FN = 5 –  ошибки FP и FN примерно одинаковы.  Но в других задачах может быть, например, что FP гораздо больше, чем FN, или наоборот.  **Анализ соотношения FP и FN важен для понимания *типа ошибок*, которые делает модель, и для выбора стратегии улучшения модели.**

**Confusion Matrix –  это основа для расчета многих других метрик классификации (Precision, Recall, F1-score, ROC AUC, PR AUC), которые мы рассмотрим в следующих шагах.**  Понимание Confusion Matrix –  ключ к пониманию этих метрик!

---

**3.Presicion**

**Precision –  показывает, *какую долю из объектов, *предсказанных* моделью как *положительный класс***, на самом деле являются *положительным классом*.**  Precision отвечает на вопрос: **"Насколько мы можем доверять positive predictions модели?"** или **"Как много из *помеченных* моделью как 'спам' на самом деле являются спамом?'"**

**Формула Precision:**

```
Precision = TP / (TP + FP)
```

**Разберем формулу Precision:**

*   **TP (True Positives) –  Числитель:**  **Верно классифицированные как положительный класс.**  Это "успех" модели в обнаружении положительного класса.
*   **TP + FP (True Positives + False Positives) –  Знаменатель:**  **Все объекты, *предсказанные* моделью как положительный класс.**  Это **все "позитивные предсказания" модели**, как верные (TP), так и ошибочные (FP).

**Интуитивное понимание Precision:**

**Precision –  это "точность" или "аккуратность" *положительных предсказаний* модели.**  **Высокий Precision означает, что модель делает *мало ложно-положительных ошибок (FP)***.  То есть, когда модель говорит "это – положительный класс", она часто бывает права.

**Пример:**

Вернемся к примеру классификации спама с Confusion Matrix:

|                | Предсказанный: Спам | Предсказанный: Не-спам |
| :------------- | :-----------------: | :--------------------: |
| Истинный: Спам |        **TP = 20**       |        **FN = 5**        |
| Истинный: Не-спам |       **FP = 5**        |       **TN = 70**       |

**Расчет Precision:**

Precision = TP / (TP + FP) = 20 / (20 + 5) = 20 / 25 = **0.8 или 80%**

**В этом примере Precision = 80%.**  Это означает, что **из всех писем, которые модель *пометила как спам*, 80% действительно оказались спамом.**  Остальные 20% (FP = 5 из 25) – это "ложные срабатывания" (не-спам письма, ошибочно помеченные как спам).

---

**4. Recall (Полнота, Чувствительность, True Positive Rate)**

**Recall –  показывает, *какую долю из объектов, *действительно* относящихся к *положительному классу***, модель *верно классифицировала* как *положительный класс*.**  Recall отвечает на вопрос: **"Насколько полно модель *обнаруживает* positive cases?"** или **"Как много *реального* спама модель *поймала* из *всего* спама?"**

**Формула Recall:**

```
Recall = TP / (TP + FN)
```

**Разберем формулу Recall:**

*   **TP (True Positives) –  Числитель:**  **Верно классифицированные как положительный класс.**  Это снова "успех" модели в обнаружении положительного класса.
*   **TP + FN (True Positives + False Negatives) –  Знаменатель:**  **Все объекты, которые *на самом деле* относятся к положительному классу.**  Это **все "реальные positive cases"**, как обнаруженные моделью (TP), так и пропущенные моделью (FN).

**Интуитивное понимание Recall:**

**Recall –  это "полнота", "чувствительность" или "доля *обнаруженных* положительных случаев от *всех* положительных случаев.**  **Высокий Recall означает, что модель делает *мало ложно-отрицательных ошибок (FN)***.  То есть, модель *не пропускает* много реальных положительных случаев.

**Пример:**

Используем ту же Confusion Matrix для спама:

|                | Предсказанный: Спам | Предсказанный: Не-спам |
| :------------- | :-----------------: | :--------------------: |
| Истинный: Спам |        **TP = 20**       |        **FN = 5**        |
| Истинный: Не-спам |       **FP = 5**        |       **TN = 70**       |

**Расчет Recall:**

Recall = TP / (TP + FN) = 20 / (20 + 5) = 20 / 25 = **0.8 или 80%**

**В этом примере Recall = 80%.**  Это означает, что **из *всех* спам-писем, которые *были в наборе данных*, модель *обнаружила* 80% из них.**  Остальные 20% (FN = 5 из 25) – это спам-письма, которые модель **пропустила** и ошибочно классифицировала как не-спам.

**Важно:**

*   **Precision и Recall –  метрики, ориентированные на *положительный класс*.**  Они показывают качество предсказаний именно для положительного класса.  Для отрицательного класса есть аналогичные метрики, но Precision и Recall обычно используют именно для положительного класса, особенно в задачах, где **положительный класс является *целевым* и *более важным* для обнаружения**.
*   **Precision и Recall *торгуют* друг с другом (Precision-Recall Tradeoff).**  **Обычно, при попытке увеличить Precision, Recall *уменьшается*, и наоборот.**  Почему так происходит?  Представь себе "очень строгий" спам-фильтр, который помечает как спам только *самые очевидные* спам-письма, в которых уверен на 100%.  Такой фильтр будет иметь **высокий Precision** (почти все помеченные как спам письма действительно будут спамом), но при этом может иметь **низкий Recall** (пропускать много "менее очевидного" спама).  И наоборот, "очень "мягкий" спам-фильтр, который помечает как спам *все подозрительные* письма, даже если не уверен до конца.  Такой фильтр будет иметь **высокий Recall** (почти весь спам будет пойман), но при этом **низкий Precision** (много не-спама будет ошибочно помечено как спам - "ложные срабатывания").  **Выбор между Precision и Recall зависит от того, *какой тип ошибок более критичен* для конкретной задачи.**

**В следующем шаге мы рассмотрим, как использовать Precision и Recall вместе, и познакомимся с F1-score, которая объединяет Precision и Recall в одну метрику.**

---

**F1-score**

**F1-score (F1-мера)** –  это **гармоническое среднее между Precision и Recall.**  Звучит немного заумно, но на самом деле это довольно простая и очень **полезная метрика, когда тебе нужно учитывать *обе* метрики –  и Precision, и Recall –  и найти между ними *баланс*.**

**Почему "гармоническое среднее"?**

Обычно, когда мы хотим усреднить два числа, мы используем **арифметическое среднее**: (a + b) / 2.  Но для Precision и Recall **арифметическое среднее не очень подходит**, потому что оно **слишком "мягкое" и может "прощать" ситуации, когда одна из метрик очень низкая, а другая высокая.**

**Пример:**

Представь, у нас есть две модели:

*   **Модель 1:** Precision = 90%, Recall = 50%
*   **Модель 2:** Precision = 60%, Recall = 80%

**Арифметическое среднее:**

*   Модель 1: (90% + 50%) / 2 = 70%
*   Модель 2: (60% + 80%) / 2 = 70%

Арифметическое среднее для обеих моделей – **одинаковое (70%)!**  Но **визуально, модели *очень разные***!  Модель 1 –  очень точная (высокий Precision), но пропускает много положительных случаев (низкий Recall).  Модель 2 –  менее точная (ниже Precision), но ловит почти все положительные случаи (высокий Recall).  **Арифметическое среднее не "чувствует" эту разницу в балансе между Precision и Recall.**

**Гармоническое среднее –  более "строгое" и "чувствительное" к *дисбалансу* между Precision и Recall.**  **Оно "штрафует" модель, если одна из метрик (Precision или Recall) сильно проседает, даже если другая при этом остается высокой.**

**Формула F1-score (гармонического среднего):**

```
F1-score = 2 * (Precision * Recall) / (Precision + Recall)
```

Или, что то же самое, но более развернуто, через TP, FP, FN:

```
F1-score = 2 * TP / (2 * TP + FP + FN)
```

**Разберем формулу F1-score:**

*   **Precision и Recall –  участвуют в формуле как *множители* в числителе и как *сумма* в знаменателе.**  Это и делает F1-score гармоническим средним.
*   **F1-score всегда находится между 0 и 1 (или 0% и 100%)**.  **Чем ближе F1-score к 1 (или 100%), тем лучше модель.**  **Идеальное значение F1-score – 1**, что означает, что модель имеет и Precision = 1, и Recall = 1 (идеальная точность и полнота).  **Минимальное значение F1-score – 0.**

**Интуитивное понимание F1-score:**

**F1-score –  это метрика, которая стремится найти *баланс* между Precision и Recall.**  **Высокий F1-score достигается только тогда, когда и Precision, и Recall *достаточно высоки* и *близки друг к другу*.**  Если одна из метрик (Precision или Recall) сильно проседает, F1-score тоже будет *значительно ниже*, даже если другая метрика при этом остается высокой.

**F1-score –  особенно полезна в задачах с *несбалансированными классами***, и **когда *важны и Precision, и Recall*, и нужно найти *оптимальный баланс* между ними.**  Например, в задачах медицинской диагностики, важно и не пропустить болезнь (высокий Recall), и не ставить ложный диагноз здоровым людям (высокий Precision).  F1-score помогает оценить, насколько модель хорошо справляется с обеими этими задачами одновременно.

**Пример:**

Вернемся к нашим двум моделям:

*   **Модель 1:** Precision = 90%, Recall = 50%
*   **Модель 2:** Precision = 60%, Recall = 80%

**Расчет F1-score:**

*   **Модель 1:** F1-score = 2 \* (0.9 \* 0.5) / (0.9 + 0.5) = 0.9 / 1.4 ≈ **0.64** или **64%**
*   **Модель 2:** F1-score = 2 \* (0.6 \* 0.8) / (0.6 + 0.8) = 0.96 / 1.4 ≈ **0.69** или **69%**

**Сравнение F1-score:**

*   Для Модели 1 F1-score ≈ 64%
*   Для Модели 2 F1-score ≈ 69%

Теперь F1-score показывает, что **Модель 2 *лучше*, чем Модель 1 (F1-score выше для Модели 2).**  F1-score "увидел" разницу в балансе между Precision и Recall, которую "пропустило" арифметическое среднее.  Модель 2, хоть и имеет немного ниже Precision, чем Модель 1, но зато имеет **гораздо выше Recall**, и **F1-score оценил этот баланс как более предпочтительный**.

**Важно:**

*   **F1-score –  это метрика для *положительного класса* (как и Precision и Recall).**  Если у тебя бинарная классификация, нужно выбрать, какой класс считать "положительным", и F1-score будет оценивать качество классификации именно для этого класса.
*   **F1-score –  это *одна метрика*, которая объединяет Precision и Recall.**  Это удобно для **сравнения моделей по *общему качеству* классификации**, когда важен баланс между Precision и Recall.  Но **F1-score не дает *полной картины* качества модели**.  В некоторых задачах может быть важно смотреть на Precision и Recall **отдельно**, чтобы лучше понять, какие типы ошибок преобладают и как их можно исправить.

Превосходно!  Раз ты готов к новым вершинам, то **отправляемся в мир ROC AUC и PR AUC –  метрик для "продвинутых" ценителей классификации!** 🚀🌟

**ROC AUC (Receiver Operating Characteristic Area Under the Curve) и PR AUC (Precision-Recall Area Under the Curve) –  это метрики, которые позволяют оценить качество классификатора *в целом*, *не привязываясь к какому-то конкретному порогу классификации*.**

**Вспомним, как работают модели классификации (например, логистическая регрессия).**  Обычно они **не выдают сразу "категорию" ("спам" или "не спам"), а предсказывают *вероятность* принадлежности к *положительному классу*** (например, вероятность того, что письмо – спам).  А уже потом, чтобы получить *конкретную категорию*, мы используем **порог (threshold)**.  Например, если вероятность > 0.5, то относим к "спаму", если ≤ 0.5 – к "не-спаму".

**Проблема в том, что *выбор порога* –  это *произвольное решение*, и *качество классификации может *сильно зависеть от выбора порога*.**  Если мы изменим порог, Precision и Recall изменятся (обычно в противоположных направлениях –  как "качели Precision-Recall Tradeoff").  **Какой порог выбрать –  не всегда очевидно.**

**ROC AUC и PR AUC –  помогают нам *оценить модель *независимо от выбора порога***, посмотреть на ее качество "в целом", *на *всех возможных порогах***.  Это особенно ценно, когда:**

*   **Нужно сравнить *разные модели* между собой *без привязки к конкретному порогу*.**  Какая модель *в целом* лучше классифицирует?  ROC AUC и PR AUC дают ответ на этот вопрос.
*   **Нужно *понять, как качество модели меняется в зависимости от порога*, и *найти оптимальный порог* для конкретной задачи.**  ROC-кривая и PR-кривая (на основе которых рассчитываются AUC) позволяют **визуально исследовать *trade-off* между разными метриками (TPR и FPR для ROC, Precision и Recall для PR) при изменении порога.**

**Начнем с ROC AUC (Receiver Operating Characteristic Area Under the Curve).**

**1. ROC-кривая (Receiver Operating Characteristic curve)**

**ROC-кривая –  это график, который показывает зависимость между двумя важными характеристиками классификатора при *изменении порога классификации*:**

*   **True Positive Rate (TPR) или Recall (Чувствительность, Доля верно классифицированных положительных объектов):**  **TPR = TP / (TP + FN)**.  Показывает, **какую долю *реальных* положительных объектов модель *обнаруживает* (классифицирует как положительные) при данном пороге.**  **Чем выше TPR, тем лучше.**
*   **False Positive Rate (FPR) или Fall-out (Доля ложных срабатываний, Доля ложно-положительных объектов):**  **FPR = FP / (FP + TN)**.  Показывает, **какую долю *реальных* отрицательных объектов модель *ошибочно классифицирует* как положительные при данном пороге.**  **Чем *ниже* FPR, тем лучше.**

**ROC-кривая строится в координатах (FPR, TPR).**  Для разных порогов классификации мы рассчитываем FPR и TPR, и строим точки на графике.  Соединяя эти точки, получаем ROC-кривую.

**Идеальный классификатор будет стремиться к *верхнему левому углу* графика ROC.**  Почему?  Потому что в верхнем левом углу:

*   **TPR = 1 (100%) –  максимально возможный True Positive Rate.**  Модель обнаруживает **все** положительные объекты.
*   **FPR = 0 (0%) –  минимально возможный False Positive Rate.**  Модель **не делает *ни одной* ложной тревоги.**

**Случайный классификатор (который просто случайно угадывает классы) будет соответствовать диагональной линии на графике ROC (линия "без умения"").**  Для случайного классификатора TPR ≈ FPR на всех порогах.

**2. ROC AUC (Area Under the Curve - Площадь под ROC-кривой)**

**ROC AUC –  это *площадь под ROC-кривой*.**  Это **численная мера качества классификатора, которая *агрегирует* информацию со *всех возможных порогов* классификации.**

**Интуитивное понимание ROC AUC:**

**ROC AUC можно интерпретировать как *вероятность того, что классификатор *ранжирует* случайно выбранный *положительный* объект *выше*, чем случайно выбранный *отрицательный* объект.**

**Чем *больше* площадь под ROC-кривой (ROC AUC), тем *лучше* классификатор.**

*   **ROC AUC = 1:**  **Идеальный классификатор.**  Все положительные объекты ранжированы выше всех отрицательных.  ROC-кривая идет в **верхний левый угол.**
*   **ROC AUC = 0.5:**  **Случайный классификатор.**  Нет различия между ранжированием положительных и отрицательных объектов.  ROC-кривая – **диагональная линия.**
*   **ROC AUC < 0.5:**  Классификатор работает **хуже, чем случайное угадывание** (хотя такое редко бывает на практике, обычно можно "перевернуть" предсказания и получить ROC AUC > 0.5).

**ROC AUC –  удобная метрика для *сравнения разных классификаторов* между собой.**  Чем выше ROC AUC, тем лучше классификатор *в целом*.  **ROC AUC *менее чувствительна* к *несбалансированности классов*, чем Accuracy, Precision, Recall и F1-score.**  Поэтому ROC AUC часто используют для оценки моделей в задачах с несбалансированными классами.

Что показывает: Качество разделения классов моделью при изменении порога классификации.

Оси:

X: False Positive Rate (FPR) — доля ложных срабатываний.

Y: True Positive Rate (TPR, Recall) — доля верно предсказанных положительных объектов.

Изменение при пороге: При снижении порога TPR и FPR растут, при повышении — уменьшаются.

ROC AUC
Что это: Площадь под ROC-кривой, мера качества модели независимо от порога.

Интерпретация:

1: Идеальное разделение классов.

0.5: Модель работает как случайное угадывание.

< 0.5: Модель хуже случайного угадывания.

Преимущества ROC AUC
Независимость от порога: Оценивает модель на всех порогах.

Устойчивость к дисбалансу классов: Полезна при несбалансированных данных.

**PR AUC (Precision-Recall Area Under the Curve) –  как и ROC AUC, оценивает качество классификатора *в целом*, *не завися от порога*, но делает это **немного по-другому, и *фокусируется на *Precision* и *Recall***, а не на TPR и FPR, как ROC AUC.**

**PR AUC –  особенно полезна в задачах с *сильно несбалансированными классами*, когда *положительный класс –  редкий, но *очень важен* для обнаружения**, и когда нас больше интересует **качество предсказаний именно для *положительного класса*** (Precision и Recall), а не общая "различительная способность" классификатора (как в ROC AUC).

**1. PR-кривая (Precision-Recall curve)**

**PR-кривая –  это график, который показывает зависимость между *Precision* и *Recall* при *изменении порога классификации*.**

**PR-кривая строится в координатах (Recall, Precision).**  Для разных порогов классификации мы рассчитываем Precision и Recall, и строим точки на графике.  Соединяя эти точки, получаем PR-кривую.

**Идеальный классификатор будет стремиться к *верхнему правому углу* графика PR.**  Почему?  Потому что в верхнем правом углу:

*   **Precision = 1 (100%) –  максимально возможная Precision.**  **Все объекты, помеченные как положительные, *действительно* являются положительными.**  Нет "ложных тревог" (FP = 0).
*   **Recall = 1 (100%) –  максимально возможный Recall.**  **Все *реальные* положительные объекты *обнаружены* и классифицированы верно.**  Нет "пропущенных целей" (FN = 0).

**Сравнение ROC-кривой и PR-кривой:**

*   **ROC-кривая (Receiver Operating Characteristic):  (FPR, TPR)**.  **Ось X – FPR (False Positive Rate), Ось Y – TPR (True Positive Rate).**  Показывает *trade-off* между **ложно-положительными срабатываниями (FP)** и **верными обнаружениями положительных объектов (TP)**, нормированными относительно *отрицательных* и *положительных* классов соответственно.  **Больше "смотрит" на качество классификации *обоих классов* (и положительного, и отрицательного).**

*   **PR-кривая (Precision-Recall):  (Recall, Precision).**  **Ось X – Recall, Ось Y – Precision.**  Показывает *trade-off* между **точностью положительных предсказаний (Precision)** и **полнотой обнаружения положительных объектов (Recall)**.  **Больше "смотрит" на качество классификации *только *положительного класса***, и **игнорирует *истинно-отрицательные срабатывания (TN)***.

**Почему PR-кривая бывает *лучше* ROC-кривой для несбалансированных классов?**

В задачах с **сильно несбалансированными классами**, когда **отрицательный класс *намного больше* положительного**, **ROC-кривая может давать *оптимистичную* оценку качества модели, даже если модель плохо работает с *положительным классом***.**  Почему так происходит?

**FPR (ось X для ROC-кривой) = FP / (FP + TN).**  **Если отрицательный класс *очень большой* (TN >> FP), то FPR будет *очень маленьким*, даже если количество FP достаточно велико в *абсолютном выражении*.**  ROC-кривая может выглядеть "хорошей" (близкой к верхнему левому углу) просто за счет того, что FPR остается низким из-за огромного TN, даже если TPR при этом не очень высокий.

**PR-кривая *не зависит* от TN и FPR, и фокусируется только на TP, FP и FN, то есть на *положительном классе и ошибках, связанных с ним*.**  **Precision = TP / (TP + FP), Recall = TP / (TP + FN).**  **PR-кривая *более чувствительна* к качеству классификации *именно *положительного класса***, и **лучше показывает различия между моделями в условиях сильного дисбаланса классов, особенно когда важна *точность и полнота обнаружения *положительного класса***.

**2. PR AUC (Area Under the Curve - Площадь под PR-кривой)**

**PR AUC –  это *площадь под PR-кривой*.**  Это **численная мера качества классификатора, которая *агрегирует* информацию с *PR-кривой* и показывает *среднее значение Precision на разных уровнях Recall*.**

**Интуитивное понимание PR AUC:**

**PR AUC можно интерпретировать как *средневзвешенное значение Precision, взвешенное по значениям Recall*.**  **Чем *больше* площадь под PR-кривой (PR AUC), тем *лучше* классификатор.**

*   **PR AUC = 1:**  **Идеальный классификатор (в смысле Precision и Recall).**  PR-кривая идет в **верхний правый угол.**
*   **PR AUC чем выше, тем лучше.**  Чем больше PR AUC, тем лучше баланс между Precision и Recall, и тем выше качество классификации *положительного класса*.
*   **PR AUC для случайного классификатора зависит от соотношения классов.**  Для сбалансированных классов PR AUC ≈ 0.5, но **для *несбалансированных* классов PR AUC для случайного классификатора может быть *значительно ниже 0.5***, особенно если положительный класс очень редкий.  **Поэтому, в задачах с редкими положительными классами, *даже не очень высокое значение PR AUC может быть *значимым улучшением* по сравнению со случайным классификатором*.**

**PR AUC –  особенно ценна для оценки моделей в задачах с *сильно несбалансированными классами*, когда *положительный класс –  редкий, но *очень важен* для обнаружения и когда важен *баланс между Precision и Recall именно для положительного класса*.**  Например, в задачах обнаружения мошенничества, редких заболеваний, таргетированной рекламы, когда **пропустить *положительный случай* (мошенничество, болезнь, заинтересованного клиента) –  гораздо хуже, чем допустить *ложную тревогу*.**
