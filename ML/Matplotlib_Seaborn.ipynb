## Matplotlib

````python
import matplotlib.pyplot as plt
import numpy as np
````

#### **Простой линейный график**

Предположим, у нас есть простые данные: координаты X и Y для нескольких точен

````python
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 1, 3, 5]
````

Чтобы построить линейный график этих данных, нам нужно всего несколько строчек кода Matplotlib:

````python
plt.plot(x_values, y_values) # Функция plot() строит линейный графие по данным x и y
plt.show() # Функция show() показывает график на экране
````

* ``` plt.plot(x_values, y_values)``` - **Основная команда для построения графиков в Matlplotlib.** Функция ```plot()``` принимает на вход два основных аргумента:
    * **Первый аргумент (x_values)** - это данные для оси X.

    * **Второй аргумент (y_values)** - это данные для оси Y.

    Функция ```plot()``` создаёт **объект Line2D** (линию), который представляет собой наш линейный график, и добавляет его к текущей области рисования (Axes).
    Если области рисования ещё нет, Matlpotlib автоматически создаст Figure и Axes.
* ```plt.show()``` - **Команда для отображения графика на экране.** Эта функция "завершает" процесс рисования и показывает все созданные графики в отдельных окнах или встраивет их в вывод Jupyter Notebook.

#### **Настройка элементов графика в Matplotlib**

* **Заголовок графика (Title):** Чтобы показать что именно изображено на графике.
* **Подписи к осям (Axis labels):** Чтобы указать величины отложены по осям X и Y,и в каких единицах измерения.
* **Цвет и стиль линнии (Line color and style):** Чтобы сдлеать графие более различимым и выразительным.

**1\. Добавление заголовка и подписей осей:**    

Функции для доваления заголовка и подписей к осям:
* **```plt.title('Заголовок графика')```:** Добавляет заголовок к текущему графику.
* **```plt.xlabel('Подпсиь оси X)```:** Добавляет подпись к оси X.
* **```plt.ylabel('Подпись оси Y)```:** Добавляет подпсиь к оси Y.

Возьмём наши предыдущие данные и добавим к графику заголовок и подпись к осям.

````python
plt.plot(x_values, y_values)

plt.title('Зависимсть Y от X')
plt.xlabel('Значения X')
plt.ylabel('Значения Y')

plt.show()
````

**2\. Настройка цвета и стиля линии:**  

Функция `plt.plot` позволяет настроивать цвет и стиль линии с помощью дополнтильный аргументов. 
* **`color='цвет'`:** Задает цвет линии. Цвет можно задавать разными способами, но стандартный - использовать **навзвания цветов**
* **`linestyle='стиль'` (`ls='стиль'`):** Задаёт стиль линнии. Основные стили:
    * `'-'` или `'solid'` – сплошная линия (по умолчанию).
    * `'--'` или `'dashed'` – пунктирная линия.
    * `':'` или `'dotted'` – точечная линия.
    * `'-.'` или `'dashdot'` – штрих-пунктирная линия.
    * `'`' или `'None'` – без линии (только точки, если есть маркеры).

````python
plt.plot(x_values, y_values, color='red', ls='dashed')

plt.title('Зависимсть Y от X')
plt.xlabel('Значения X')
plt.ylabel('Значения Y')

plt.show()
````

#### **Точечные диаграммы**

В отличие от линейных графиков, которые показывают **линию тренда** или **зависимость** между переменными, **точечные диаграммы показывают отдельные точки данных**.

**Когда используют точечные диаграммы?**
* **Для иследования взаимосвязи между элементами:** Точечные диаграммы отлично подходят для визуализации того, как одна переменная меняется в зависимости от другой. Можно увидеть, есть ли какая-то закономерность, тренд, или точки распределены случайно.
* **Для выявления кластеров и выбросов:** На точечной диаграмме часто хорошо видны группы точек (кластеры) и точки, которые сильно отличаются от остальных (выбросы).
* **Когда важны отдельные точки данных:** В отличие от линейных графиков, где линия как бы "сглаживает" данные, точечные диаграммы показывают каждую точку данных индивидуально, что может быть важно, если каждая точка представляет собой отдельное наблюдение или событие.
* **Для визаулизации многомерных данных:** Точечные диаграммы можно использовать для отображения не только двух, но и большего количества переменных, если использовать цвет, размер и форму точек для кодирования дополнительных измерений.

Для создания точеченых диаграмм в Matplotlib используется функция **`plt.scatter()`**   

**Основное синтаксис **`plt.scatter(x, y, ...)`**:
* `x` **- Данные для оси X**
* `y` **- Данные для оси Y**
* **Дополнительные аргументы (параметры):** Позволяют настроить внешний вид точек: цвет, размер, форму маркера и т.д.

**Простейшая точечная диаграмма:**

````python
plt.scatter(x_values, y_values) 

plt.title('Точечная диаграмма зависимости Y от X')
plt.xlabel('Значения X')
plt.ylabel('Значения Y')

plt.show()
````

**Настройка внешнего вида точек в `plt.scatter()`:**

Функция `plt.scatter()` предоставляет множество параметров для настройки внешнего вида точек.  Рассмотрим основные:

*   **`marker='маркер'`:**  Задает **форму маркера** (точки).  Основные маркеры:
    *   `'o'` – **кружок** (по умолчанию).
    *   `'.'` – **точка**.
    *   `','` – **пиксель**.
    *   `'x'` – **крестик**.
    *   `'+'` – **плюс**.
    *   `'*'` – **звездочка**.
    *   `'s'` – **квадрат**.
    *   `'d'` – **ромб**.
    *   `'^'` – **треугольник вверх**.
    *   `'v'` – **треугольник вниз**.
    *   `<` – **треугольник влево**.
    *   `'>'` – **треугольник вправо**.
    *   `'p'` – **пятиугольник**.
    *   `'h'` – **шестиугольник 1**.
    *   `'H'` – **шестиугольник 2**.

*   **`s=размер`:**  Задает **размер маркера** в **пунктах в квадрате**.  Чем больше число, тем больше размер точек.

*   **`c=цвет` или `color=цвет`:**  Задает **цвет маркера**.  Цвет можно задавать так же, как и для линий в `plt.plot()` (названия цветов, HEX-коды, RGB-значения и т.д.).

````python
plt.scatter(x_values, y_values, color='purple', marker='x', s=100)

plt.title('Точечная диаграмма (крестики, фиолетовые, большие)')
plt.xlabel('Значения X')
plt.ylabel('Значения Y')

plt.show()
````

#### **Столбчатые диаграммы**   

**Столбчатые диаграммы (bar charts)** – это один из самых интуитивно понятных и широко используемых типов графиков. Их часто называют гистограммами, но важно различать столбчатые диаграммы (bar charts) и гистограммы (histograms). Столбчатые диаграммы используются для сравнения значений между разными категориями, а гистограммы – для отображения распределения непрерывных данных.

**Когда используются столбчатые диаграммы?**

* **Для сравнения категориальных данных:** Когда у тебя есть категории (например, страны, продукты, отделы продаж) и ты хочешь сравнить какую-то числовую характеристику для каждой категории (например, ВВП стран, продажи продуктов, прибыль отделов).
* **Для отображения рейтингов и сравнений:** Столбчатые диаграммы отлично подходят для визуализации рейтингов, например, рейтингов популярности фильмов, оценок товаров, мест в соревнованиях.
* **Для показа изменений во времени для разных категорий:** Можно использовать группированные столбчатые диаграммы, чтобы сравнить, как меняется какая-то величина во времени для разных категорий.

Для создания столбчатых диаграмм в Matplotlib используется функция **`plt.bar()`**.

**Основной синтаксис `plt.bar(x, height, ...)`:** 
* `x`**- Позиции столбцов на оси X.** Обычно это **категориальные значения** (например названия категорий). Matplotlib автоматически располагает стобцы на оси X.
* ``height``**- Высоты столбцов.** Это **числовые значения** которые нужно сравнить между категориями.
* **Дополнительные аргументы (параметры):**  Позволяют настроить внешний вид столбцов: цвет, ширину, выравнивание и т.д.

**Простейшая столбчатая диаграмма:**

Предположим, у нас есть данные о продажах по разным категориям товаров:

````python
categories = ['Категория А', 'Категория Б', 'Категория В', 'Категория Г']
sales = [150, 220, 180, 250]

plt.bar(categories, sales)

plt.title('Продажи по категориям')
plt.xlabel('Категория товаров')
plt.ylabel('Объем продаж')

plt.show()
````

**Настройка внешнего вида столбцов в ```plt.bar```():**

* **`color='цвет'`:** Задает **цвет столбцов**.
* **`width='ширина'`:** Задает **ширину столбцов**. Значение от 0 до 1 (по умолчанию 0.8).
* **`align='выравнивание'`:** Задает **выравнивание столбцов по оси X**. 
    * `'center'` - **центр столбца** (по умолчанию).
    * `'edge'` - **левый край столбца**.
* **`edgecolor=цвет_границы`:** Задает **цвет границы столбцов.**
* **`linewidth=толщина_границы`:** Задает **толщину границы столбцов**.

````python
plt.bar(categories, sales, color='grey', edgecolor='black', width=0.5, linewidth=10)

plt.show()
````

#### **Гистограммы**

**Ключевое различие между столбчатыми диаграммами (bar charts) и гистограммами (histograms):**

* **Столбчатые диаграммы (Bar Charts):**
    * Используются для **сравнения значений между разными категориями**.
    * **Ось X – категориальная:** названия категорий (например, 'Категория А', 'Категория Б', 'Город 1', 'Город 2', 'Яблоко', 'Банан').
    * **Ось Y – числовая:** значения, которые сравниваются для каждой категории (например, 'Продажи', 'Население', 'Цена').
    * **Между столбцами есть промежутки**, так как категории дискретны и не связаны между собой в порядке следования по оси X.
* **Гистограммы (Histograms):**
    * Используются для отображения **распределения непрерывных данных**.
    * **Ось X – непрерывная:** диапазон значений непрерывной переменной (например, 'Возраст', 'Рост', 'Температура', 'Время'). Ось X разбивается на интервалы (bins), и гистограмма показывает, сколько значений попадает в каждый интервал.
    * **Ось Y – частота (количество) или относительная частота (доля) значений, попадающих в каждый интервал**.
    * **Столбцы идут вплотную друг к другу** (обычно), потому что ось X непрерывная, и интервалы (bins) следуют друг за другом без разрывов.

**Грубо говоря:**

* **Столбчатая диаграмма:** "Сколько чего-то в каждой категории?" (сравнение категорий)
* **Гистограмма:** "Как часто встречаются разные значения непрерывной переменной?" (распределение значений)

Для создания гистограмм в Matplotlib используется функция **```plt.hist()```**

````python
data = np.random.randn(1000) # генерируем 1000 случайных чисел из стадартного распределения

plt.hist(data)

plt.title('Гистограмма распределения случайных данных')
plt.xlabel('Значения данных')
plt.ylabel('Частота')

plt.show()
````

**Настройка внешнего вида гистограммы в `plt.hist()`:**

Функция `plt.hist()` также предоставляет параметры для настройки гистограммы.  Рассмотрим основные:

*   **`bins=количество_интервалов` или `bins=границы_интервалов`:**  Задает **количество интервалов (bins)** или **границы интервалов**.
    *   Если задать **целое число** (например, `bins=20`), то Matplotlib разобьет данные на указанное количество **равных по ширине** интервалов.
    *   Если задать **список или массив** (например, `bins=[ -3, -2, -1, 0, 1, 2, 3 ]`), то это будут **границы интервалов**, и интервалы могут быть разной ширины.

*   **`color=цвет`:**  Задает **цвет столбцов гистограммы**.

*   **`edgecolor=цвет_границы`:**  Задает **цвет границы столбцов гистограммы**.

*   **`alpha=прозрачность`:**  Задает **прозрачность столбцов**.  Значение от 0 (полностью прозрачный) до 1 (полностью непрозрачный).  Полезно, если нужно наложить несколько гистограмм друг на друга.

*   **`histtype='тип'`:**  Задает **тип гистограммы**.  Основные типы:
    *   `'bar'` – **обычная столбчатая гистограмма** (по умолчанию).
    *   `'step'` – **ступенчатая гистограмма** (только контур столбцов, без заливки).
    *   `'stepfilled'` – **ступенчатая гистограмма с заливкой**.

````python
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

plt.title('Гистограмма (30 интервалов, голубая, с черной границей, полупрозрачная)')
plt.xlabel('Значения данных')
plt.ylabel('Частота')

plt.show()
````

## Seaborn

**Ключевые преимущества Seaborn:**

*   **Высокий уровень абстракции:**  Seaborn предоставляет **более высокоуровневый интерфейс** для создания графиков, чем Matplotlib.  Многие вещи, которые в Matplotlib нужно делать "руками", в Seaborn делаются автоматически или с помощью нескольких параметров.  Это **упрощает и ускоряет** процесс визуализации.
*   **Специализация на статистической визуализации:**  Seaborn изначально разработан для создания **статистических графиков**.  В нем "из коробки" есть множество функций для построения распределений, взаимосвязей, сравнений, которые часто используются в анализе данных и машинном обучении.
*   **Интеграция с Pandas DataFrame:**  Seaborn **отлично работает с Pandas DataFrame**.  Ты можешь передавать DataFrame и имена столбцов прямо в функции Seaborn, и Seaborn сам "разберется", как отобразить данные на графике.  Это делает визуализацию данных из Pandas **очень удобной и эффективной**.
*   **Красивый внешний вид по умолчанию:**  Графики Seaborn **выглядят привлекательно "из коробки"**, даже без дополнительной настройки.  Seaborn использует **разумные стили и цветовые палитры**, что делает визуализации более профессиональными и понятными.
*   **Расширение возможностей Matplotlib:**  Seaborn **не заменяет Matplotlib, а дополняет его**.  Seaborn построен на основе Matplotlib, и ты можешь использовать **все возможности Matplotlib для тонкой настройки графиков Seaborn**, если это необходимо.  То есть, изучив Matplotlib, ты уже знаешь "фундамент" Seaborn, и тебе будет легче его освоить.

````python
import seaborn as sns 
````

Для начала построим точечную диаграмму scatter plot.    
Вот как это делалось в Matplotlib:

````python
import matplotlib.pyplot as plt

x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 1, 3, 5]

plt.scatter(x_values, y_values)
plt.show()
````

А вот как посторить аналогичную диаграмму в Seaborn:

````python
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 1, 3 ,5]

sns.scatterplot(x=x_values, y=y_values)

plt.show()
````

**Разница в коде Seaborn:**

*   `sns.scatterplot(x=x_values, y=y_values)` –  **Функция `sns.scatterplot()` в Seaborn – это аналог `plt.scatter()` в Matplotlib, но более высокоуровневая.**
    *   Аргументы `x` и `y` в `sns.scatterplot()` принимают **данные для оси X и Y**.  Здесь мы также передаем списки `x_values` и `y_values`.
    *   В Seaborn **мы передаем данные *по именам аргументов* (`x=`, `y=`)**, а не просто по порядку, как в `plt.scatter()`.  Это делает код более читаемым и позволяет передавать больше параметров в произвольном порядке.

#### **Seaborn и Pandas**   

Seaborn понимает структуру данных DataFrame и позволяет указывать:
* DataFrame, из которого брать данные.
* Имена столбцов для осей X и Y.
* Столбы для кодирования доп. измерений (цвет, размер, форма точек).

Точечная диаграмма с использованием Pandas Dataframe:

````python
import pandas as pd

data_dict = {
    'Возраст': [25, 30, 28, 35, 22, 40, 29, 33, 45, 27],
    'Доход': [30000, 50000, 45000, 60000, 25000, 70000, 48000, 55000, 80000, 40000]
}

df_seaborn = pd.DataFrame(data_dict)
print('DataFrame для Seaborn:')
df_seaborn.head()
````

````python
sns. scatterplot(data=df_seaborn, x='Возраст', y='Доход')

plt.title('Зависимость дохода от возраста (Seaborn + DataFrame)')
plt.xlabel('Возраст (годы)')
plt.ylabel('Доход')

plt.show()
````

**Пояснение:**

*   `sns.scatterplot(data=df_seaborn, x='Возраст', y='Доход')` –  **Вот как просто строится точечная диаграмма в Seaborn с использованием DataFrame!**
    *   `data=df_seaborn` –  **Аргумент `data` указывает, какой DataFrame использовать в качестве источника данных.**  Мы передаем наш DataFrame `df_seaborn`.
    *   `x='Возраст'` –  **Аргумент `x` указывает, какой столбец DataFrame использовать для оси X.**  Мы просто указываем **имя столбца** `'Возраст'` **в виде строки**!  Seaborn сам найдет этот столбец в DataFrame `df_seaborn` и возьмет его значения для оси X.
    *   `y='Доход'` –  **Аналогично, аргумент `y` указывает столбец `'Доход'` для оси Y.**

#### **Кодирование дополнительных измерений на графике**

Точечные диаграммы, которые мы строили до сих пор, показывали взаимосвязь только между **двумя переменными** (X и Y).  Но часто в данных у нас есть **больше, чем две переменные**, и мы хотим увидеть, как они взаимодействуют друг с другом, и как **дополнительные переменные влияют на взаимосвязь между X и Y**.

**Seaborn позволяет нам кодировать дополнительные измерения на 2D-графиках, используя визуальные признаки, такие как:**

*   **Цвет точек (hue):**  Разные цвета для точек, принадлежащих к разным категориям.
*   **Размер точек (size):**  Разный размер точек в зависимости от значения третьей числовой переменной.
*   **Форма точек (style):**  Разная форма точек для разных категорий.

**Сегодня мы начнем с самого распространенного и мощного способа кодирования дополнительного измерения – с цвета точек (hue).**

**Параметр `hue` в Seaborn**

Параметр `hue` (англ. "оттенок", "цвет") – это **ключевой параметр в Seaborn для добавления *категориального* измерения к графику.**  Когда ты используешь `hue`, Seaborn **автоматически раскрашивает точки (или другие элементы графика) в разные цвета в зависимости от значений в *категориальном столбце* DataFrame.**

**Как использовать `hue` в `sns.scatterplot()`:**

Просто добавь аргумент `hue='имя_категориального_столбца'` в функцию `sns.scatterplot()`.  Seaborn сам позаботится о выборе подходящей цветовой палитры и создании легенды, объясняющей, какой цвет соответствует какой категории.

**Пример: Точечная диаграмма с кодированием категорий цветом (hue)**

````python
data_dict_hue = {
    'Возраст': [25, 30, 28, 35, 22, 40, 29, 33, 45, 27],
    'Доход': [30000, 50000, 45000, 60000, 25000, 70000, 48000, 55000, 80000, 40000],
    'Пол': ['Жен', 'Муж', 'Муж', 'Муж', 'Жен', 'Муж', 'Жен', 'Жен', 'Муж', 'Жен'] # Категориальный столбец 'Пол'
}

df_seaborn_hue = pd.DataFrame(data_dict_hue)
display(df_seaborn_hue.head())
````

````python
sns.scatterplot(data=df_seaborn_hue, x='Возраст', y='Доход', hue='Пол')

plt.title('Зависимость дохода от возраста с разделением по полу (hue="Пол")')
plt.xlabel('Возраст (годы)')
plt.ylabel('Доход (у.е.)')

plt.show()
````

**Пояснение:**

*   `sns.scatterplot(data=df_seaborn_hue, x='Возраст', y='Доход', hue='Пол')` –  **Мы добавили всего один аргумент – `hue='Пол'`!**
    *   `hue='Пол'` –  **Указывает Seaborn, что нужно использовать столбец 'Пол' для кодирования цвета точек.**  Seaborn автоматически определит уникальные значения в столбце 'Пол' ('Жен' и 'Муж') и назначит **разные цвета** для точек, соответствующих каждой категории.  Также Seaborn автоматически добавит **легенду**, которая объясняет, какой цвет какому полу соответствует.

#### **Box Plots**

**Ящичковые диаграммы (Box Plots)** – это отличный способ **визуально сравнить распределения *числовой* переменной для разных *категорий***.  В отличие от точечных диаграмм, которые показывают отдельные точки, ящичковые диаграммы показывают **сводную статистику распределения** для каждой категории: медиану, квартили, размах, выбросы.

**Когда используются ящичковые диаграммы?**

*   **Для сравнения распределений между группами:**  Когда ты хочешь сравнить, как распределена числовая переменная (например, доход, возраст, оценка) в разных категориях (например, по полу, по типу продукта, по региону).  Ящичковые диаграммы позволяют быстро увидеть различия в центральной тенденции, разбросе и форме распределения между группами.
*   **Для выявления выбросов:**  Ящичковые диаграммы наглядно показывают выбросы – значения, которые выходят за пределы "нормального" размаха данных.
*   **Для сравнения нескольких распределений на одном графике:**  Ящичковые диаграммы компактны и позволяют эффективно сравнивать несколько распределений рядом друг с другом.

**Что показывает ящичковая диаграмма?**

![Box Plot Anatomy](https://upload.wikimedia.org/wikipedia/commons/1/1a/Boxplot_vs_PDF.svg)
_(Изображение с Википедии: [https://en.wikipedia.org/wiki/Box_plot](https://en.wikipedia.org/wiki/Box_plot))_

*   **Ящик (Box):**  Ограничен **первым квартилем (Q1)** и **третьим квартилем (Q3)**.  Внутри ящика лежат **50% "центральных" данных**.  Длина ящика (Q3 - Q1) – это **межквартильный размах (IQR)**, который характеризует разброс "центральной" половины данных.
*   **Медиана (Median, средняя линия внутри ящика):**  Показывает **"середину" распределения**.  Делит данные пополам.
*   **Усы (Whiskers):**  Линии, идущие от ящика.  Обычно простираются до **минимального значения, не меньшего, чем Q1 - 1.5 * IQR**, и до **максимального значения, не большего, чем Q3 + 1.5 * IQR**.  Усы показывают "нормальный" размах данных, исключая выбросы.
*   **Выбросы (Outliers, точки за пределами усов):**  Точки, которые лежат **за пределами усов**.  Считаются **аномально большими или маленькими значениями**.

**Как создать ящичковую диаграмму в Seaborn?**

Для создания ящичковых диаграмм в Seaborn используется функция **`sns.boxplot()`**.

**Основной синтаксис `sns.boxplot(data=DataFrame, x='категориальный_столбец', y='числовой_столбец', ...)`:**

*   **`data=DataFrame`:**  **DataFrame, содержащий данные.**
*   **`x='категориальный_столбец'`:**  **Столбец DataFrame с *категориальными* значениями**, категории для сравнения.  Будет отображаться на оси X.
*   **`y='числовой_столбец'`:**  **Столбец DataFrame с *числовыми* значениями**, распределение которого мы хотим сравнить между категориями.  Будет отображаться на оси Y.
*   **Дополнительные аргументы (параметры):**  Для настройки внешнего вида, добавления цветового кодирования (например, `hue`) и т.д.

**Пример: Ящичковая диаграмма Seaborn для сравнения дохода по полу**

````python
sns.boxplot(data=df_seaborn_hue, x='Пол', y='Доход', hue='Пол')

plt.title('Распределение дохода по полу (Box Plot)')
plt.xlabel('Пол')
plt.ylabel('Доход (у.е.)')

plt.show()
````

#### **Гистограммы**

**Основной синтаксис `sns.histplot(data=DataFrame, x='числовой_столбец', ...)`:**

*   **`data=DataFrame`:**  **DataFrame, содержащий данные.**
*   **`x='числовой_столбец'`:**  **Столбец DataFrame с *непрерывными* числовыми значениями**, для которого нужно построить гистограмму.  Будет отображаться на оси X (интервалы значений).
*   **Дополнительные аргументы (параметры):**  Для настройки внешнего вида, количества интервалов (bins), цвета, добавления кривой плотности распределения и т.д.

**Настройка внешнего вида гистограммы в `sns.histplot()`:**

Функция `sns.histplot()` также предоставляет параметры для настройки гистограммы, **аналогичные `plt.hist()` в Matplotlib**, но с более "Seaborn-ориентированным" синтаксисом (например, цвет задается через `color=`, а не просто позиционным аргументом).  Вот некоторые полезные параметры:

*   **`bins=количество_интервалов`:**  Задает **количество интервалов**.  Работает так же, как и в `plt.hist()`.

*   **`color=цвет`:**  Задает **цвет столбцов гистограммы**.

*   **`edgecolor=цвет_границы`:**  Задает **цвет границы столбцов гистограммы**.

*   **`alpha=прозрачность`:**  Задает **прозрачность столбцов**.

*   **`kde=True` или `kde=False`:**  **Добавить или не добавлять линию оценки плотности ядра (Kernel Density Estimate, KDE)** поверх гистограммы.  **KDE – это сглаженная оценка формы распределения.**  `kde=True` (добавить KDE) – часто бывает очень полезно, чтобы лучше увидеть общую форму распределения, особенно если данных не очень много и гистограмма получается "рваной".  `kde=False` (не добавлять KDE, по умолчанию) – только столбчатая гистограмма.


````python
sns.histplot(data=df_seaborn_hue, x='Возраст', bins=5, color='skyblue', edgecolor='black', alpha=1, kde=False)

plt.title('Распределение возраста (гистограмма Seaborn)')
plt.xlabel('Возраст (годы)')
plt.ylabel('Частота')

plt.show()
````

#### **Скрипипчные диограммы**

**В чем особенность скрипичных диаграмм?**

*   **Сочетание боксплота и KDE:**  Скрипичная диаграмма – это как **гибрид** ящичковой диаграммы и гистограммы с линией KDE.  В центре скрипичной диаграммы обычно рисуется **мини-боксплот** (ящик с медианой и квартилями), а **по бокам** от него "разрастается" **симметричная фигура, напоминающая скрипку**, которая представляет собой **отзеркаленную оценку плотности ядра (KDE)** распределения данных.
*   **Более детальное представление распределения, чем боксплот:**  Если ящичковая диаграмма показывает только сводную статистику (квартили, медиану, выбросы), то скрипичная диаграмма дает **более полное представление о форме распределения** – видны "пики", "хвосты", модальность и другие особенности распределения, благодаря KDE.
*   **Эстетически привлекательные:**  Скрипичные диаграммы выглядят очень **эффектно и выразительно**, особенно при сравнении нескольких распределений рядом друг с другом.

**Когда используются скрипичные диаграммы?**

*   **Для сравнения распределений между группами (как и боксплоты, но более детально):**  Когда нужно не просто сравнить "средние значения" или "разброс", а **увидеть *форму* распределения** числовой переменной в разных категориях и сравнить эти формы.  Например, сравнить распределение оценок студентов в разных группах, распределение времени отклика сайта в зависимости от типа запроса и т.д.
*   **Когда важна форма распределения:**  Если тебе важно понять, является ли распределение нормальным, скошенным, бимодальным, или имеет какие-то другие особенности – скрипичная диаграмма поможет это увидеть.
*   **В дополнение к боксплотам:**  Скрипичные диаграммы часто используют **вместе с боксплотами**, чтобы получить как сводную статистику (боксплот в центре), так и детальное представление о форме распределения (скрипка по бокам).

**Как создать скрипичную диаграмму в Seaborn?**

Для создания скрипичных диаграмм в Seaborn используется функция **`sns.violinplot()`**.

**Основной синтаксис `sns.violinplot(data=DataFrame, x='категориальный_столбец', y='числовой_столбец', ...)`:**

Синтаксис **полностью аналогичен `sns.boxplot()`**:

*   **`data=DataFrame`:**  **DataFrame, содержащий данные.**
*   **`x='категориальный_столбец'`:**  **Столбец DataFrame с *категориальными* значениями**, категории для сравнения.  Будет отображаться на оси X.
*   **`y='числовой_столбец'`:**  **Столбец DataFrame с *числовыми* значениями**, распределение которого мы хотим сравнить между категориями.  Будет отображаться на оси Y.
*   **Дополнительные аргументы (параметры):**  Для настройки внешнего вида, добавления цветового кодирования (например, `hue`), изменения внутреннего представления и т.д.

````python
sns.violinplot(data=df_seaborn_hue, x='Пол', y='Доход')

plt.title('Распределение дохода по полу (Violin Plot)')
plt.xlabel('Пол')
plt.ylabel('Доход (у.е.)')

plt.show()
````

**Настройка внешнего вида скрипичных диаграмм в `sns.violinplot()`:**

Функция `sns.violinplot()` предоставляет много параметров для настройки внешнего вида.  Вот некоторые интересные:

*   **`color=цвет` или `palette=палитра_цветов`:**  Задает **цвет скрипок** (или **цветовую палитру** для разных категорий, если используется `hue`).  Можно использовать названия цветов, HEX-коды, цветовые палитры Seaborn (например, `'pastel'`, `'muted'`, `'deep'`, `'colorblind'`, и другие - погугли "seaborn color palettes").

*   **`inner='тип_внутреннего_отображения'`:**  Задает **тип внутреннего отображения внутри "скрипки"**.  Возможные значения:
    *   `'box'` – **мини-боксплот** (по умолчанию).
    *   `'quartile'` – **квартили** (три линии, показывающие Q1, медиану, Q3).
    *   `'point'` – **отдельные точки** данных внутри скрипки (может быть полезно для небольших наборов данных).
    *   `'stick'` – **"палочки"** на линии медианы, показывающие плотность данных.
    *   `None` – **без внутреннего отображения** (только форма скрипки).

*   **`scale='метод_масштабирования'`:**  Задает **масштабирование ширины скрипок**.  Влияет на то, насколько широкими будут скрипки для разных категорий.  Возможные значения:
    *   `'area'` – **площадь скрипки** пропорциональна количеству наблюдений в категории (по умолчанию).  Общая площадь всех скрипок на графике будет примерно одинаковой.
    *   `'count'` – **ширина скрипки** пропорциональна количеству наблюдений в категории.  Более многочисленные категории будут иметь более широкие скрипки.
    *   `'width'` – **все скрипки имеют одинаковую ширину**.

````python
sns.violinplot(data=df_seaborn_hue, x='Пол', y='Доход', palette='pastel', inner='quartile')
plt.title('Распределение дохода по полу (Violin Plot) с настройками')
plt.xlabel('Пол')
plt.ylabel('Доход (у.е.)')
plt.show()
````
