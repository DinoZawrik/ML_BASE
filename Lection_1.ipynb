### 1. NumPy

Зачем нужен NumPy - он написан на C, что делает его операции очень быстрыми по сравнению с обычными циклами в python. В ml Постоянно надо работаеть с векторами и матрицами, а numpy предоставляет структуру данных 'ndarray', Которая идеально подходит для работы с векторами и матрицами. Массивы numpy позволяют работать только с данными одного типа (int или float), но зато мы получаем куда большую скорость выполнения.

Создаём массив NumPy
Самый просто способ - это преобразовать обычный список python в массив numpy с помошью функции np.array()

````python
import numpy as np # NumPy принято сокращать как np для краткости
python_list = [1, 2, 3, 4, 5]
numpy_list = np.array(python_list)

print(python_list)
print(numpy_list)
print(type(numpy_list))
````

Ключевые характеристика массивов Numpy:     
1. __Однородность типа данных:__ Все элементы должны быть одного типа данных. Это может быть и __int__ и __float__ и __bool__ и многое другое. Проверть тип данных можно с помощью атрибута __dtype__
```python
    print(numpy_list.dtype)
```
