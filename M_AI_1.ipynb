
## Задача 17
**1. Лингвистические переменные:**

* **Входные переменные:**
    * **Нагрузка на компьютер (Load):**  Характеризует текущую загрузку компьютера в процентах.
        * Термы (T_Load): {"Низкая", "Средняя", "Высокая"}
        * Универсальное множество (X_Load): [0, 100] (проценты)
    * **Размер задачи (TaskSize):**  Относительный размер задачи.
        * Термы (T_TaskSize): {"Маленькая", "Средняя", "Большая"}
        * Универсальное множество (X_TaskSize): [1, 10] (условные единицы)
    * **Производительность компьютера (Performance):**  Относительная производительность компьютера.
        * Термы (T_Performance): {"Низкая", "Средняя", "Высокая"}
        * Универсальное множество (X_Performance): [1, 10] (условные единицы)

* **Выходная переменная:**
    * **Приоритет компьютера (Priority):**  Определяет приоритет компьютера для выполнения новой задачи (от 0 до 1).
        * Термы (T_Priority): {"Низкий", "Средний", "Высокий"}
        * Универсальное множество (X_Priority): [0, 1]

**2. Функции принадлежности:**

* **Для (Load):**
    * **"Низкая":**  Трапеция [0, 0, 30, 50].
    * **"Средняя":**  Треугольник [30, 50, 70].
    * **"Высокая":**  Трапеция [50, 70, 100, 100].

* **Для (TaskSize):**
    * **"Маленькая":** Трапеция, [1, 1, 3, 5]
    * **"Средняя":** Треугольник, [3, 5, 7]
    * **"Большая":** Трапеция, [5, 7, 10, 10]

* **Для (Performance):**
    * **"Низкая":** Трапеция, [1, 1, 3, 5]
    * **"Средняя":** Треугольник, [3, 5, 7]
    * **"Высокая":** Трапеция, [5, 7, 10, 10]

* **Для (Priority):**
    * **"Низкий":** Трапеция, [0, 0, 0.3, 0.5]
    * **"Средний":** Треугольник, [0.3, 0.5, 0.7]
    * **"Высокий":** Трапеция, [0.5, 0.7, 1, 1]

**3. База нечетких правил:**

1. **ЕСЛИ Нагрузка на компьютер = "Низкая" И Размер задачи = "Большая" ТО Приоритет компьютера = "Высокий"**
2. **ЕСЛИ Нагрузка на компьютер = "Высокая" ТО Приоритет компьютера = "Низкий"**
3. **ЕСЛИ Производительность компьютера = "Высокая" И Размер задачи = "Большая" ТО Приоритет компьютера = "Высокий"**
4. **ЕСЛИ Производительность компьютера = "Низкая" ТО Приоритет компьютера = "Низкий"**
5. **ЕСЛИ Нагрузка на компьютер = "Средняя" И Размер задачи = "Средняя" ТО Приоритет компьютера = "Средний"**
6. **ЕСЛИ Нагрузка на компьютер = "Низкая" И Размер задачи = "Маленькая" ТО Приоритет компьютера = "Средний"**
7. **ЕСЛИ Нагрузка на компьютер = "Средняя" И Производительность компьютера = "Средняя" ТО Приоритет компьютера = "Средний"**
8. **ЕСЛИ Производительность компьютера = "Высокая" И Нагрузка на компьютер = "Средняя" ТО Приоритет компьютера = "Средний"**
9. **ЕСЛИ Производительность компьютера = "Средняя" И Размер задачи = "Большая" ТО Приоритет компьютера = "Средний"**
10. **ЕСЛИ Нагрузка на компьютер = "Высокая" И Производительность компьютера = "Низкая" ТО Приоритет компьютера = "Низкий"**


**4. Проверка на полноту (качественная):**

База правил выглядит достаточно полной, так как учтены разные комбинации нагрузки, размеры задачи и производительность для определения приоритета. Есть правила для "Низкого", "Среднего" и "Высокого" приоритета.  Также, для каждой входной переменной и каждого ее терма есть правила, где они используются. 

**5. Пример нечеткого вывода:**

Предположим, у нас есть компьютер со следующими характеристиками:

* **Нагрузка на компьютер (Load):** 40%
* **Размер задачи (TaskSize):** 7
* **Производительность компьютера (Performance):** 6

**Фаззификация:**

* **Для Load = 40%:**
    * μ_Низкая(40%) ≈ 0.25
    * μ_Средняя(40%) ≈ 0.5
    * μ_Высокая(40%) ≈ 0
* **Для TaskSize = 7:**
    * μ_Маленькая(7) = 0
    * μ_Средняя(7) = 0
    * μ_Большая(7) ≈ 0.75
* **Для Performance = 6:**
    * μ_Низкая(6) = 0
    * μ_Средняя(6) ≈ 0.5
    * μ_Высокая(6) ≈ 0.5

**Вычисление степеней уверенности правил и аккумуляция:**

* **Правило 1:** мин(μ_Низкая(40%), μ_Большая(7)) = мин(0.25, 0.75) = 0.25  -> Приоритет = "Высокий" (степень 0.25)
* **Правило 2:** μ_Высокая(40%) = 0 -> Приоритет = "Низкий" (степень 0)
* **Правило 3:** мин(μ_Высокая(6), μ_Большая(7)) = мин(0.5, 0.75) = 0.5 -> Приоритет = "Высокий" (степень 0.5)
* **Правило 4:** μ_Низкая(6) = 0 -> Приоритет = "Низкий" (степень 0)
* **Правило 5:** мин(μ_Средняя(40%), μ_Средняя(7)) = мин(0.5, 0) = 0 -> Приоритет = "Средний" (степень 0)
* **Правило 6:** мин(μ_Низкая(40%), μ_Маленькая(7)) = мин(0.25, 0) = 0 -> Приоритет = "Средний" (степень 0)
* **Правило 7:** мин(μ_Средняя(40%), μ_Средняя(6)) = мин(0.5, 0.5) = 0.5 -> Приоритет = "Средний" (степень 0.5)
* **Правило 8:** мин(μ_Высокая(6), μ_Средняя(40%)) = мин(0.5, 0.5) = 0.5 -> Приоритет = "Средний" (степень 0.5)
* **Правило 9:** мин(μ_Средняя(6), μ_Большая(7)) = мин(0.5, 0.75) = 0.5 -> Приоритет = "Средний" (степень 0.5)
* **Правило 10:** мин(μ_Высокая(40%), μ_Низкая(6)) = мин(0, 0) = 0 -> Приоритет = "Низкий" (степень 0)

**Аккумуляция:**

* **"Низкий" приоритет:** max(0, 0, 0) = 0
* **"Средний" приоритет:** max(0, 0, 0.5, 0.5, 0.5) = 0.5
* **"Высокий" приоритет:** max(0.25, 0.5) = 0.5

**Результирующее нечеткое множество для "Приоритет компьютера":**

* "Низкий" - степень принадлежности 0
* "Средний" - степень принадлежности 0.5
* "Высокий" - степень принадлежности 0.5

**Вывод:**

Для компьютера с нагрузкой 40%, задачей размера 7 и производительностью 6, нечеткая логическая система определила "Средне-Высокий" приоритет.

#### Реализация на Python

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# 1. Лингвистические переменные
# Входные переменные
load = ctrl.Antecedent(np.arange(0, 101, 1), 'load')  # Нагрузка, 0-100%
task_size = ctrl.Antecedent(np.arange(1, 11, 1), 'task_size') # Размер задачи, 1-10
performance = ctrl.Antecedent(np.arange(1, 11, 1), 'performance') # Производительность, 1-10

# Выходная переменная
priority = ctrl.Consequent(np.arange(0, 1.01, 0.01), 'priority') # Приоритет, 0-1

# 2. Функции принадлежности
# Для нагрузки (load)
load['низкая'] = fuzz.trapmf(load.universe, [0, 0, 30, 50])
load['средняя'] = fuzz.trimf(load.universe, [30, 50, 70])
load['высокая'] = fuzz.trapmf(load.universe, [50, 70, 100, 100])

# Для размера задачи (task_size)
task_size['маленькая'] = fuzz.trapmf(task_size.universe, [1, 1, 3, 5])
task_size['средняя'] = fuzz.trimf(task_size.universe, [3, 5, 7])
task_size['большая'] = fuzz.trapmf(task_size.universe, [5, 7, 10, 10])

# Для производительности (performance)
performance['низкая'] = fuzz.trapmf(performance.universe, [1, 1, 3, 5])
performance['средняя'] = fuzz.trimf(performance.universe, [3, 5, 7])
performance['высокая'] = fuzz.trapmf(performance.universe, [5, 7, 10, 10])

# Для приоритета (priority)
priority['низкий'] = fuzz.trapmf(priority.universe, [0, 0, 0.3, 0.5])
priority['средний'] = fuzz.trimf(priority.universe, [0.3, 0.5, 0.7])
priority['высокий'] = fuzz.trapmf(priority.universe, [0.5, 0.7, 1, 1])

#Визуализация функций принадлежности
load.view()
task_size.view()
performance.view()
priority.view()
```

```python
# 3. База нечетких правил
rule1 = ctrl.Rule(load['низкая'] & task_size['большая'], priority['высокий'])
rule2 = ctrl.Rule(load['высокая'], priority['низкий'])
rule3 = ctrl.Rule(performance['высокая'] & task_size['большая'], priority['высокий'])
rule4 = ctrl.Rule(performance['низкая'], priority['низкий'])
rule5 = ctrl.Rule(load['средняя'] & task_size['средняя'], priority['средний'])
rule6 = ctrl.Rule(load['низкая'] & task_size['маленькая'], priority['средний'])
rule7 = ctrl.Rule(load['средняя'] & performance['средняя'], priority['средний'])
rule8 = ctrl.Rule(performance['высокая'] & load['средняя'], priority['средний'])
rule9 = ctrl.Rule(performance['средняя'] & task_size['большая'], priority['средний'])
rule10 = ctrl.Rule(load['высокая'] & performance['низкая'], priority['низкий'])


# 4. Система управления нечетким выводом
priority_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10])
priority_simulation = ctrl.ControlSystemSimulation(priority_ctrl)

# 5. Входные значения для примера
load_input = 40
task_size_input = 7
performance_input = 6

priority_simulation.input['load'] = load_input
priority_simulation.input['task_size'] = task_size_input
priority_simulation.input['performance'] = performance_input

# 6. Вычисление нечеткого вывода
priority_simulation.compute()

# 7. Вывод результатов
print("Результаты для:")
print(f"  Нагрузка: {load_input}%")
print(f"  Размер задачи: {task_size_input} у.е.")
print(f"  Производительность: {performance_input} у.е.")
print(f"Приоритет компьютера: {priority_simulation.output['priority']:.2f}") # Выводим приоритет

# Визуализация результата
priority.view(sim=priority_simulation)
load.view(sim=priority_simulation)
task_size.view(sim=priority_simulation)
performance.view(sim=priority_simulation)
```
