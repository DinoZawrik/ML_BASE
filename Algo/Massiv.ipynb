## Python инструментарий

**Списки (`list`) - основная структура для массивов в Python.**

*   Тип `list` - это упорядоченная, изменяемая коллекция элементов.
*   В Python списки динамические, могут менять размер.
*   Элементы могут быть разных типов (но для задач LeetCode лучше однотипные).

**Создание списков:**

*   **Литерал списка `[]`:**
    ```python
    пустой_список = []
    числа = [1, 2, 3]
    строки = ['элемент', 'второй']
    ```
*   **Конструктор `list()`:**
    ```python
    из_кортежа = list((1, 2))
    из_строки = list("строка")
    из_диапазона = list(range(5))
    ```
*   **List comprehensions** (рассмотрим ниже).

**Индексация и срезы:**

*   **Индексация `arr[i]`:** доступ к элементу по индексу (с 0).
    ```python
    список = [10, 20, 30]
    первый_элемент = список[0]  # 10
    последний_элемент = список[-1] # 30
    ```
*   **Срезы `arr[start:end:step]`:** извлечение подсписка.
    ```python
    список = [10, 20, 30, 40, 50]
    подсписок = список[1:4]    # [20, 30, 40]
    начало = список[:3]       # [10, 20, 30]
    конец = список[2:]        # [30, 40, 50]
    шаг = список[::2]       # [10, 30, 50]
    разворот = список[::-1]    # [50, 40, 30, 20, 10]
    ```

**Методы списков:**

*   **Изменение списка (in-place):**
    *   `.append(элемент)`: добавить в конец.
    *   `.extend(итерируемый)`: добавить несколько элементов.
    *   `.insert(индекс, элемент)`: вставить по индексу.
    *   `.remove(элемент)`: удалить первое вхождение.
    *   `.pop(индекс=-1)`: удалить и вернуть элемент (последний).
    *   `.clear()`: очистить список.
    *   `.reverse()`: развернуть.
    *   `.sort()`: сортировать (по возрастанию).

    ```python
    список = [1, 2, 3]
    список.append(4)      # [1, 2, 3, 4]
    список.insert(0, 0)   # [0, 1, 2, 3, 4]
    удаленный = список.pop() # 4 (список: [0, 1, 2, 3])
    ```

*   **Получение информации (не меняют список):**
    *   `len(список)`: длина списка.
    *   `.count(элемент)`: количество вхождений.
    *   `.index(элемент)`: индекс первого вхождения.

    ```python
    список = [1, 2, 2, 3]
    длина = len(список)    # 4
    количество_2 = список.count(2) # 2
    индекс_3 = список.index(3)   # 3
    ```

*   **Копирование:**
    *   `.copy()`: создать копию списка.

    ```python
    копия_списка = список.copy()
    ```

**List comprehensions (генераторы списков):**

*   **Синтаксис:** `[выражение for элемент in итерируемый_объект if условие]`.
*   Компактное создание списков.
*   Часто эффективнее циклов.

    ```python
    квадраты = [x**2 for x in range(5)] # [0, 1, 4, 9, 16]
    четные = [x for x in range(10) if x % 2 == 0] # [0, 2, 4, 6, 8]
    матрица = [[0 for _ in range(2)] for _ in range(2)] # [[0, 0], [0, 0]]
    ```

## Теория (ключевые моменты) - Массивы

**1. Быстрый доступ по индексу (O(1)):**

*   Массивы хранят элементы **непрерывно** в памяти.
*   Это обеспечивает **константное время доступа** к элементу по индексу `arr[i]`.
*   Важное преимущество массивов (Python lists) перед, например, связными списками.

**2. Временная сложность операций:**

*   Важно знать **временную сложность** основных операций над массивами (списками Python).
*   Понимание сложности помогает выбирать **эффективные алгоритмы** и **оптимизировать код**.

| Операция                                  | Временная сложность (Средняя) | Временная сложность (Худшая) | Примечания                                                                                                                                                                                                   |
| :---------------------------------------- | :---------------------------: | :---------------------------: | :----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Доступ по индексу** `arr[i]`             |       **O(1)**              |       **O(1)**              | Константное время. Один из самых быстрых способов доступа к данным.                                                                                                                                           |
| **Вставка в конец** `arr.append(x)`        |   **O(1) (амортизированное)**   |       **O(n)**              | В среднем добавление в конец - очень быстро.  Худший случай (редко) - когда нужно перевыделить память под список большего размера и скопировать все существующие элементы.                                |
| **Вставка в середину/начало** `arr.insert(i, x)` |       **O(n)**              |       **O(n)**              | Линейное время.  Нужно сдвинуть все элементы, начиная с позиции вставки, вправо. Чем ближе к началу вставка, тем больше элементов нужно сдвигать.                                                              |
| **Удаление из конца** `arr.pop()`           |       **O(1)**              |       **O(1)**              | Константное время.                                                                                                                                                                                            |
| **Удаление из середины/начала** `arr.pop(i)`, `del arr[i]` |       **O(n)**              |       **O(n)**              | Линейное время.  Нужно сдвинуть все элементы, начиная с позиции удаления, влево, чтобы заполнить "дыру".  Чем ближе к началу удаление, тем больше элементов нужно сдвигать.                               |
| **Удаление по значению** `arr.remove(x)`    |       **O(n)**              |       **O(n)**              | Линейное время.  В худшем случае нужно просмотреть весь список, чтобы найти первое вхождение элемента `x`.  После нахождения и удаления также может потребоваться сдвиг элементов (как при `pop(i)`).      |
| **Поиск элемента** `x in arr`, `arr.index(x)`, `arr.count(x)` |       **O(n)**              |       **O(n)**              | Линейное время. В худшем случае нужно просмотреть весь список, чтобы найти элемент (или убедиться, что его нет, или посчитать все вхождения).                                                              |
| **Сортировка** `arr.sort()`, `sorted(arr)`    |     **O(n log n)**            |     **O(n log n)**            | Квазилинейное время. Используется алгоритм Timsort, который очень эффективен на практике и хорошо работает на частично отсортированных данных.                                                              |
| **Разворот** `arr.reverse()`              |       **O(n)**              |       **O(n)**              | Линейное время. Нужно поменять местами примерно половину элементов списка.                                                                                                                                   |
| **Копирование (срезы, `arr.copy()`, `list(arr)`)** |       **O(n)**              |       **O(n)**              | Линейное время. Нужно скопировать все элементы из исходного списка в новый.                                                                                                                              |
| **`len(arr)`**                               |       **O(1)**              |       **O(1)**              | Получение длины списка - константное время. Python списки хранят информацию о своей длине.                                                                                                                   |
| **`arr.clear()`**                             |       **O(n)**              |       **O(n)**              | Линейное время. Нужно пройтись по всем элементам и освободить память, которую они занимали.                                                                                                                |

**3. Два указателя и скользящее окно:**

*   **Два указателя:**  Использование двух индексов для скоординированного обхода массива.  Вариации:
    *   С разных концов (для отсортированных массивов, палиндромов).
    *   Быстрый и медленный (для удаления/разделения элементов).
*   **Скользящее окно:**  Окно (подмассив), "скользящее" по массиву.  Для поиска оптимальных подмассивов/подстрок.
*   Часто позволяют снизить сложность до **O(n)** и экономят память **O(1)**.

**4. Префиксные суммы:**

*   Техника для **быстрого вычисления сумм подмассивов**.
*   Создается массив `prefix_sum`, где `prefix_sum[i]` - сумма элементов `arr[0...i-1]`.
*   Сумма `arr[left...right]` вычисляется за **O(1)** как `prefix_sum[right+1] - prefix_sum[left]`.
*   Полезно, когда нужно **много раз** считать суммы подмассивов.
*   Предобработка - **O(n)**, запросы - **O(1)**.
